### Tutorial Wahrscheinlichkeitsrechnung I


In der Wahrscheinlichkeitsrechnung geht es um das rechnen mit sogenannten **Wahrscheinlichkeiten**. 

### Was ist eine Wahrscheinlichkeit ?

Eine Wahrscheinlichkeit ist eine Zahl zwischen 0 und 1 (bzw. 0% und 100%), die den Grad der Unsicherheit ausdrückt ob ein Ereignis eintrifft oder nicht, bzw. den Grad der Unsicherheit ausdrückt ob eine bestimmte Information wahr oder falsch ist.

#### Warum sollte ich Wahrscheinlichkeitsrechnung nutzen ?

Wahrscheinlichkeitsrechnung ist, das kann man mathematisch beweisen, die einzige Methode mit der man, wenn man über unsichere oder unvollständige Informationen verfügt, völlig rationale Entscheidungen treffen kann bei denen man nicht von jemandem unfair über das Ohr gehauen werden kann der nicht über mehr Informationen verfügt als man selbst.

Wenn man sich über alle Informationen sicher ist dann sind die Ergebnisse der Wahrscheinlichkeitsrechnung identisch zu den Ergebnissen klassischer Logik.

#### Was heisst hier Ereignis ?

Mit Ereignis kann zum Beispiel das Ereignis gemeint sein, dass ein Würfel bei einem bestimmten Würfelwurf eine 1 zeigt, dass ein Pfeil eine Scheibe trifft. 

In jedem Fall ist ein Ereignis etwas das entweder eintrifft oder nicht. Es gibt kein Dazwischen. 

Darüber ob es eintrifft oder bereits eingetroffen ist kann man unsicher sein. Das kann der Fall sein wenn man über unvollständige Informationen verfügt oder wenn es sich um ein Ereignis handelt das noch in der Zukunft liegt, und dessen Ausgang vom Zufall bestimmt wird (wie ein idealisierter Würfelwurf).

#### Interpretationen einer Wahrscheinlichkeit

Es gibt mehrere Möglichkeiten, einen Wahrscheinlichkeitswert, der ja zunächst nur eine Zahl ist, zu interpretieren. 

 * **Frequentistische Interpretation**: Die Wahrscheinlichkeit kann als eine Aussage über den verlauf hypothetischer zukünftiger Zufallsexperimente interpretiert werden. In diesem Fall bedeutet eine Erfolgswahrscheinlichkeit von 50% bei einem Experiment die Vorhersage, dass wenn man das Experiment **unendlich oft** wiederholen würde, 50% aller Versuche Erfolg hätten. 

 * **Bayesianische Interpretation**: Man kann eine Wahrscheinlichkeit auch als einen Prozentwert interpretieren der den Grad des eigenen Glaubens (bzw. die eigene Unsicherheit) darüber angibt ob eine bestimmte Information wahr ist, oder ein Ereignis eintreten wird. 

 * **Multiple-Welten Interpretation**: Die Wahrscheinlichkeit ist eine Zahl die den Anteil möglicher (also nicht  völlig auszuschliessender) Welten angibt in denen ein Ereignis eintritt.
 
 * **Mathematische Interpretation**: Es ist nur eine Zahl die bestimmten Einschränkungen und axiomatischen Rechenregeln gehorcht. Weitere Interpretationen gehen zu weit.

Alle diese Interpretationen ergeben in bestimmten Situationen Sinn, und in anderen nicht. Deshalb sollte man sie kennen. Alle Interpretationen sind auf philosophischer Ebene kontrovers.


### Die Axiome der Wahrscheinlichkeitsrechnung

Das folgende sind Eigenschaften der Wahrscheinlichkeitsrechnung die sozusagen vorausgesetzt werden müssen, damit das ganze funktioniert, und die Eigenschaft hat dass man nicht übers Ohr gehauen werden kann:

1.) Eine Wahrscheinlichkeit ist eine Zahl zwischen 0 und 1 (inklusiv)

2.) Die Summe der Wahrscheinlichkeit aller möglichen sich jeweils gegenseitig ausschliessenden Ereignisse ist 1

3.) Können zwei Ereignisse A *und* B nicht gleichzeitig wahr sein, dann gilt:
Die Wahrscheinlichkeit dass A *oder* B wahr ist, beträgt genau die Wahrscheinlichkeit von A plus die Wahrscheinlichkeit von B

### Was bedeutet P(A), P(x=3) oder P(A,B) ?

Für die Wahrscheinlichkeit eines Ereignisses verwendet man die Notation $ P(A) $. D.h. $ P(A) $ ist die Wahrscheinlichkeit dass Ereignis A (was auch immer das ist. Es ist an dieser Stelle ein Platzhalter) eintrifft.

Manchmal schreibt man auch etwas wie $ P(x=3) $. Damit ist dann die Wahrscheinlichkeit gemeint dass $ x = 3 $ ist. Entsprechend kann man auch schreiben $ P(x<3) $ für die Wahrscheinlichkeit dass P kleiner 3 ist. 

$ x $ hingegen könnte zum Beispiel das Ergebnis eines Würfelwurfs sein.

Hier ist wichtig zu beachten:

 * Wenn $ x $ das Ergebnis eines Würfelwurfs ist, könnte es theoretisch eine beliebige Anzahl von Werten annehmen.
 * Die Aussage $ x = 3 $ kann hingegen nur wahr oder falsch sein. Daher ist es ein Ereignis wenn es wahr ist. (Und es wäre ein anderes Ereignis wenn es falsch ist).

Die Notation $ P( A , B) $ bedeutet die Wahrscheinlichkeit, dass sowohl A als auch B wahr sind (man kann es also als $ P (\text{A und B} )$ lesen.

#### Was ist die Gegenwahrscheinlichkeit $ P(\hat{A}) $ ?

Aus den Axiomen folgt sofort, dass wenn P(A) einen bestimmten Wert hat, dann muss das Gegenteil (also das Ereignis, dass das Ereignis A *nicht* eintritt) genau 1-P(A) betragen. 

Wenn wir mit $ P(\hat{A}) $ die Wahrscheinlichkeit bezeichnen dass A **nicht** eintritt, dann gilt:

$$ P (\hat{A}) = 1 - P(A) $$

**Warum ?** Weil ein Ereignis entweder wahr ist oder nicht. Deshalb muss die Wahrscheinlichkeit dass es wahr ist oder nicht zusammengenommen gleich 1 sein (Axiom 2). 

Daraus folgt: $P(A) + P(\hat{A}) = 1 $ und daraus folgt $ P(\hat{A}) = 1 - P(A) $

### Was bedeutet P(A | B) oder P(x=3 | B) ?

Man verwendet die Notation $ P(A | B ) $ um den verbleibenden Grad der Unsicherheit anzugeben, ob ein Ereignis A wahr ist, wenn man schon weiss dass das Ereignis B wahr ist. Das nennt man eine *bedingte* Wahrscheinlichkeit, und man kann es als die "Wahrscheinlichkeit von A gegeben B" lesen. 

### Rechenregeln für bedingte Wahrscheinlichkeiten

Die Wahrscheinlichkeit dass A und B gleichzeitig wahr sind, ist gleich hoch wie die Wahrscheinlichkeit dass A wahr ist, multipliziert mit der bedingten Wahrscheinlichkeit dass B wahr ist, wenn A wahr ist.

Oder anders ausgedrückt (Merke: $ P(A,B) $ ist die Wahrscheinlichkeit dass sowohl A als auch B eintreten): 

$$
P(A,B) = P(A | B ) \cdot P(B) = P(B | A) \cdot P(A)
$$

### Unabhängige Ereignisse

Wenn die Wahrscheinlichkeit zweier Ereignisse A und B absolut nichts miteinander zu tun hat, nennt man sie unabhängig.

Unabhängigkeit von Ereigissen bedeutet insbesondere, dass die Wahrscheinlichkeit von A gegeben B gleich der Wahrscheinlichkeit von A ist. D.h. das Wissen ob B wahr ist oder nicht hat keinen Einfluss auf den Grad meiner Unsicherheit ob das Ereignis A eintrifft oder nicht.

In der Notation von oben würde das also heissen:

$$ P( A | B ) = P ( A ) $$

und

$$ P( B | A ) = P ( B ) $$

#### Warum ist das so ?

Das ist sozusagen die Definition dessen, was es bedeutet dass zwei Ereignisse unabhängig sind.

#### Kombinierte Wahrscheinlichkeit unabhängiger Ereignisse

Daraus folgt übrigens insbesondere, dass wenn A und B voneinander unabhängige Ereignisse sind, die Wahrscheinlichkeit dass beide eintreten einfach berechnet werden kann indem man ihre einzelnen Wahrscheinlichkeiten multipliziert:

$$
P(A,B) = P(A | B ) \cdot P(B) = P( A ) \cdot P(B)
$$

#### Beispiele für unabhängige Ereignisse

 * Die Wahrscheinlichkeit dass bei zwei hintereinander geworfenen Würfeln zunächst (A) eine 1 und dann beim zweiten Wurf (B) eine 6 fällt. 
 * Die Wahrscheinlichkeit aus einer Urne mit 10 blauen Kugeln und 10 schwarzen Kugeln zunächst (A) eine blaue und dann (B) eine schwarze zu zählen, **wenn nach dem ersten Wurf die Kugel wieder zurückgelegt wird**.
 
#### Beispiele für abhängige Ereignisse

 * Die Wahrscheinlichkeit dass bei einem geworfenen Würfel, (A) eine 1 fällt, oder (B) eine 6 fällt. (Sie schliessen sich gegenseitig aus).
 * Die Wahrscheinlichkeit aus einer Urne mit 10 blauen Kugeln und 10 schwarzen Kugeln zunächst (A) eine blaue und dann (B) eine schwarze zu zählen, **wenn nach dem ersten Wurf die Kugel nicht wieder zurückgelegt wird**.
 * Die Wahrscheinlichkeit (A) dass der Boden auf der Strasse nass ist, und (B) dass der nächste Passant auf der Strasse einen Regenschirm trägt. 
 * Die Wahrscheinlichkeit: (A) bei einem ersten Würfelwurf (W6) mindestens eine 5 zu würfeln, und die Wahrscheinlichkeit (B) dass wenn man zum ersten Würfelwurf (W6) einen weiteren Würfelwurf addiert, das Ergebnis grösser oder gleich 9 ist.

#### Beispiel

Nehmen wir an, wir wollen zwei mal hintereinander eine Münze werfen. Wenn wir uns jetzt vorstellen wie sich die Zukunft entfalten kann, dann verzweigt sie sich mit jeder Entscheidung die fällt...

## Die Binomialverteilung

Die sogenannte Binomialverteilung ist eine mathematische Funktion, welche die Wahrscheinlichkeit berechnet, dass man genau **k** Erfolge erzielt, wenn man **N** unabhängige Zufallsexperimente macht bei denen jeweils mit *immer der gleichen Wahrscheinlichkeit* **p** nur eines von 2 möglichen Ergebnissen herauskommen kann (entweder Erfolg oder Misserfolg).

Die Variablen **k** (Anzahl der Erfolge), **N** (Anzahl der Experimente) und **p** (Erfolgswahrscheinlichkeit eines einzelnen Experiments) sind dabei die Parameter der Funktion.

Es versteht sich von selbst dass **k** eine ganze Zahl zwischen 0 und **N** sein muss, da man wenn man **N** Experimente macht, man auch maximal **N** Erfolge erzielen kann. 

Aus diesem Grund kann man einfach für gegebenes **N** und gegebenes **p** ein die Werte als Diagramm ausgebenwelches die Wahrscheinlichkeit angibt dass **k** Erfolge erzielt werden.


#### Wie funktioniert diese Funktion ?

Im Grunde genommen funktioniert sie genau so wie oben in dem Beispiel mit den Würfelwürfen. Es werden alle möglichen Pfade und ihre jeweiligen Einzelwahrscheinlichkeiten ermittelt. Und am Ende werden die einzelnen Wahrscheinlichkeiten aller Pfade die zu genau **k** Erfolgen führen aufaddiert. Und das ist dann das Ergebnis.

Es gibt im wesentlichen nur 2 Unterschiede zum Beispiel oben:

 * Es gibt jeweils nur 2 mögliche Ergebnisse ( Erfolg oder Misserfolg)
 * Die Erfolgswahrscheinlichkeit beträgt **p**, die Misserfolgswahrscheinlichkeit jeweils **1-p**
 

### Beispiel: Berechnung der Binomialverteilung für N=4 und p=0.2

Unten kann man die Felder z.B. so lesen: Experiment 1 (E1) hat einen Erfolg (E1=1) oder einen Misserfolg (E1=0).
Am Ende werden die Erfolge gezählt und die entsprechenden Wahrscheinlichkeiten aufaddiert.

Das ganze kann natürlich für beliebig grosse N und beliebige p zwischen 0 und 1 berechnet werden. Nur lässt es sich dann nicht mehr vernünftig grafisch darstellen.

```python
from matplotlib import pylab as plt
import numpy as np
from scipy.stats import binom as binomial_distribution
%matplotlib inline
```

```python
def plot_binomial(p,N):
    k = np.array(range(0,N+1), dtype=np.int32)
    pk = binomial_distribution( n=N,p=p).pmf(k)

    plt.Figure(figsize=(16,16))
    plt.bar(k, pk, 1.0)
    plt.xlabel('k Erfolge')
    if (N<=20):
        plt.xticks(k+0.5, k)
    else:
        plt.xticks(np.array(list(range(0,N+1, N//16)))+0.5,np.array(list(range(0,N+1, N//16))))
    plt.ylabel('Wahrscheinlichkeit')
    plt.title("%d Experimente mit jeweils Wahrscheinlichkeit %0.2f%%" % (N, 100.0*p))
    plt.show()

plot_binomial(0.5, 10)
plot_binomial(0.05, 10)
plot_binomial(0.995, 16)
plot_binomial(0.5, 20)
plot_binomial(0.5, 40)
plot_binomial(0.1, 100)
plot_binomial(0.04, 100)
```


## Ein Konfidenzintervall für die Binomialverteilung

Das ganze ist soweit schön und gut, allerdings kann man aus der Binomialverteilung noch nicht direkt ein Konfidenzintervall für den Wert p ablesen. 

**Warum ?**

Weil wir hier immer die Wahrscheinlichkeit von **k** Erfolgen bei **N** Experimenten bei *gegebener Erfolgswahrscheinlichkeit* **p** berechnen. 

### Was ist, wenn wir die Erfolgswahrscheinlichkeit p nicht kennen ?

Wenn wir p nicht kennen, können wir natürlich auch nicht die Wahrscheinlichkeit von **k** Erfolgen berechnen.
Was wir allerdings tun können, ist:

Bei gegebenem **N** und gegebenem **k** könnten wir doch theoretisch einfach mal alle plausiblen Werte für **p** durchprobieren, und schauen wie wahrscheinlich es ist, dass wir mit diesem Wert dann die beobachtete Anzahl von Erfolgen erzielen.

Da es sich bei p um eine (wenn auch unbekannte) Wahrscheinlichkeit handelt, wissen wir, dass der Wert sich zwischen 0 und 1 bewegen muss.

#### Und wo kommt das Konfidenzintervall her ?

Aus dieser Verteilung lässt sich jetzt ein Konfidenzintervall bestimmen. Das funktioniert effektiv, indem man einen zentralen Wertebereich für **p** findet, in dem sich 95% der Wahrscheinlichkeit sammelt.

Um das genaue Konfidenzintervall zu berechnen gibt es 2 Möglichkeiten:

1.) Man berechnet es numerisch. Das bedeutet effektiv, man probiert geschickt verschiedene Werte bis man mit hinreichender Genauigkeit ein Ergebnis hat.

2.) Man benutzt die von cleveren Mathematikern bereitgestellte Funktion "beta.ppf", die den gleichen Effekt mit deutlich geringerem Rechenaufwand erzielt.

Wir verwenden jetzt die zweite Variante. Zum Verständnis reicht es aber zu wissen, dass wir einen Bereich um den Wahrscheinlichsten Wert von p herum suchen, so dass:

 * 95% der Wahrscheinlichkeit (blaue Fläche) für p in diesem Bereich ist
 * Jeweils 2.5% der Wahrscheinlichkeit (graue Flächen) im Bereich vor und hinter dem Konfidenzintervall liegt.


```python

from matplotlib import pylab
import numpy as np
from scipy.stats import binom as binomial_distribution
from scipy.stats import beta
%matplotlib inline

def plot_likelihood2(k, N):
    fig = pylab.figure(figsize=(5,3))
    fig.clf()
    fig.hold(True)
    # Wir probieren einfach mal Werte für p, von 0 bis 1.0 in Schritten von 0.01 aus..
    betadist = beta(1+k, 1+N-k)
    confidence_interval = (betadist.ppf(0.025), betadist.ppf(0.975)) #betadist.interval(0.95)
    
    p = np.linspace(0, 1.0, num=201)
    p1 = p[(p<confidence_interval[0]) | (p>confidence_interval[1])]
    p2 = p[(p>=confidence_interval[0]) & (p<=confidence_interval[1])]
    likelihoods1 = betadist.pdf(p1) #binomial_distribution(n=N, p=p1).pmf(k)
    likelihoods2 =  betadist.pdf(p2) #binomial_distribution(n=N, p=p2).pmf(k)

    plt.bar(p1, likelihoods1, 1.0/len(p), align='center', edgecolor='grey', color='grey', alpha=0.6)
    plt.bar(p2, likelihoods2, 1.0/len(p), align='center', edgecolor='blue', color='blue', alpha=0.9)
    
    plt.xlabel("Erfolgswahrscheinlichkeit p")
    plt.xlim([0.0, 1.0])

    plt.ylabel("Likelihood")
    plt.title("Verteilung fuer p bei %d Erfolgen von %d Versuchen\nKonfidenzintervall [ %0.2f%%, %0.2f%% ]" % (k, N, confidence_interval[0]*100.0, confidence_interval[1]*100.0))
    plt.show()
    
plot_likelihood2(4,20)
plot_likelihood2(12,20)
plot_likelihood2(1,30)
plot_likelihood2(3,30)
plot_likelihood2(1,384)
plot_likelihood2(4,384)
plot_likelihood2(4,300)
plot_likelihood2(10,30)
plot_likelihood2(100,300)
plot_likelihood2(1000,3000)
plot_likelihood2(10000,30000)

```

Wie man an den letzten 4 Beispielen schön sehen kann, wird der Wertebereich des Konfidenzintervalls immer enger, desto mehr Versuche unternommen werden.

### Darf man das einfach so berechnen ?

Das ist eine etablierte Methode der sogenannten **bayesianischen Statistik**, die an dieser Stelle sehr viel Sinn macht.

### Bayes-Formel

Um zu begründen warum man das so berechnen können sollte, erinnern wir uns an die Formeln über bedingte Wahrscheinlichkeiten:

$$
P(A|B) \cdot P(B) = P(B|A) \cdot P(A)
$$

daraus folgt, indem man beide Seiten der Gleichung durch $P(B) $ teilt:

$$
P(A | B ) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

Das ist die sogenannte **Bayes-Formel**. 

In unserem Fall wissen wir, wie man die Wahrscheinlichkeit berechnet $K=k$ Erfolge zu generieren gegeben dass wir eine Erfolgswahrscheinlichkeit von **p** bei **N** Versuchen haben:

Wir kennen also $P(k | p)$. Was wir allerdings berechnen wollen ist $P( p | k )$. Setzen wir mal in Bayes-Formel ein:

$$
P(p | k) = \frac{P(k | p) \cdot P(p)}{P(k)}
$$

An dieser Stelle stellt sich zunächst folgende Fragen:

##### Was setze ich für P(k) ein ?

Für $P(k)$ kann 1 eingesetzt werden, da wir sicher sind dass wir k Erfolge beobachtet haben. Es ist also ein Fakt über den wir keine Unsicherheit haben.

##### Was setze ich für P(p) ein ?

Das der Punkt, der die Zweige der Statistik trennt. Man nennt $ P(p)$ den sogenannten Prior für p. Einigermassen unkontrovers ist, dass man die Daten sprechen lassen sollte, und keine Annahmen über den Wert von p treffen sollte ausser dass es sich um eine Wahrscheinlichkeit, und somit um einen Wert von 0 bis 1 handelt. Das nennt man dann einen nichtinformativen Prior.

Das schöne ist, dass wenn man weiss dass für alle P(p) der gleiche Wert eingesetzt werden soll, wir auf mathematischem Weg durch Gleichungsumformungen herausfinden können dass $ P(p) = 1 $ gilt.

#### Warum ?

Da die Summe für $ P( p | k) $ für alle möglichen Werte von p gleich 1 sein muss (Axiom 1 der Wahrscheinlichkeitsrechnung). Das ist, unter der Voraussetzung dass P(p) konstant ist, wiederum nur der Fall wenn $ P(p) = 1 $ gilt.

#### Und darf man das jetzt so rechnen ?

Ja, denn das bedeutet, dass wenn wir in diesem Fall keine weitere Annahme über p treffen als dass es sich um eine Wahrscheinlichkeit handelt, wir für P(k) und P(p) jeweils 1 einsetzen, und zum Schluss kommen dass

$$
P( p | k ) = P( k | p)
$$

Und das ist, was wir genutzt haben um die Konfidenzintervalle zu berechnen.

Man sollte darauf verweisen dass es sich hier um ein **Bayesianisches Konfidenzintervall** unter Verwendung eines nichtinformativen Priors handelt.

```python
import networkx as nx
```

```python
dice_types = [4,3]

G = nx.DiGraph()
start = 'Start'
G.add_node('Start', { 'p' : 1.0, 'label' : 'Start' } )


def build_dice_graph( G, parent, dice_types, dice_idx=1):
    result = []
    for dice_result in range(1, dice_types[0]+1):
            label = 'D%d=%d' % (dice_idx, dice_result)
            if (parent!=start):
                nodeid = '%s , %s' % (parent, label)
            else:
                nodeid = label
            
            G.add_node(nodeid, { 'dice_result' : dice_result, 'p' : G.node[parent]['p']/dice_types[0], 'label' : nodeid})
            edge = G.add_edge(parent, nodeid, {"p" : G.node[parent]['p']/dice_types[0] })
            if (len(dice_types)>1):
                result += build_dice_graph(G, nodeid, dice_types[1:], dice_idx+1)
            else:
                result += [nodeid]
    return result

def add_sum_nodes(G, leaf_nodes):
        nodes = leaf_nodes
        sums = set()
        for n in nodes:
            nsum = 0
            np = n
            visited = set([n])
            p = G.node[np]['p']
            
            while n!=start:
                result = G.node[n]['dice_result']
                nsum+=result
                in_edges = G.in_edges(n)
                if (len(in_edges)>0):
                    n2 = in_edges[0][0]
                    n = n2
            nodeid = 'Summe=%d' % (nsum)
            if (not G.has_node(nodeid)):
                G.add_node(nodeid, { 'p' : p, 'label' : nodeid})
            else:
                G.node[nodeid]['p'] = G.node[nodeid]['p'] + p
            G.add_edge(np, nodeid, {'p' : p})
            
        
leaf_nodes = build_dice_graph(G, start, dice_types)          

add_sum_nodes(G, leaf_nodes)

```

```python

G2 = nx.DiGraph()
start = 'Start'
G2.add_node('Start', { 'p' : 1.0, 'label' : 'Start' } )


def build_binomial_graph( G, parent, N, prob, dice_idx=1):
    result = []
    for dice_result in range(2):
            label = 'E%d=%d' % (dice_idx, dice_result)
            if (parent!=start):
                nodeid = '%s , %s' % (parent, label)
            else:
                nodeid = label
            pprob = G.node[parent]['p']
            pr = pprob * (dice_result * prob + (1.0-dice_result) * (1.0-prob) )
            G.add_node(nodeid, { 'dice_result' : dice_result, 'p' : pr, 'label' : nodeid})
            edge = G.add_edge(parent, nodeid, {"p" : pr })
            if (dice_idx<=N):
                result += build_binomial_graph(G, nodeid, N-1, prob, dice_idx+1)
            else:
                result += [nodeid]
    return result

def add_success_nodes(G, leaf_nodes):
        nodes = leaf_nodes
        sums = set()
        for n in nodes:
            nsum = 0
            np = n
            visited = set([n])
            p = G.node[np]['p']
            
            while n!=start:
                result = G.node[n]['dice_result']
                nsum+=result
                in_edges = G.in_edges(n)
                if (len(in_edges)>0):
                    n2 = in_edges[0][0]
                    n = n2
            nodeid = 'Erfolge=%d' % (nsum)
            if (not G.has_node(nodeid)):
                G.add_node(nodeid, { 'p' : p, 'label' : nodeid})
            else:
                G.node[nodeid]['p'] = G.node[nodeid]['p'] + p
            G.add_edge(np, nodeid, {'p' : p})
            
    
leaf_nodes2 = build_binomial_graph(G2, start, 4, 0.2, 1)          

add_success_nodes(G2, leaf_nodes2)
```

```python
%%javascript


function plot_prob_graph(graph, target) {

    var margin = {top: 1, right: 1, bottom: 6, left: 1};
    var _width = 960 - margin.left - margin.right;
    var _height = 500 - margin.top - margin.bottom;
    var width=_width;
    var height=_height;
    window.margin = margin;
    window.width=width;
    window.height=height;
    var formatNumber = d3.format(",.1f"),
        format = function(d) { return formatNumber(d) + ""; },
        color = d3.scale.category20();

    var svg = d3.select(target).append("svg")
        .attr("width", _width + margin.left + margin.right)
        .attr("height", _height + margin.top + margin.bottom)
      .append("g")
        .attr("transform", "translate(" + margin.left + "," + margin.top + ")");

    var sankey = d3.sankey()
        .nodeWidth(15)
        .nodePadding(10)
        .size([_width, _height]);

    var path = sankey.link();
    console.log(path);


    function plot_sankey(energy) {

      for (var i=0;i<energy.nodes.length;i++) {
        energy.nodes[i].value = energy.nodes[i].p;
      }

      for (var i=0;i<energy.links.length;i++) {
        energy.links[i].value = energy.links[i].p;
      }
      
      sankey
          .nodes(energy.nodes)
          .links(energy.links)
          .layout(32);

      var link = svg.append("g").selectAll(".link")
          .data(energy.links)
        .enter().append("path")
          .attr("class", "link")
          .attr("d", path)
          .style("stroke-width", function(d) { return Math.max(1, d.dy); })
          .sort(function(a, b) { return b.dy - a.dy; });

      link.append("title")
          .text(function(d) { return d.source.label + " → " + d.target.label + "\n" + format(100.0*d.p) + "%"; });

      var node = svg.append("g").selectAll(".node")
          .data(energy.nodes)
        .enter().append("g")
          .attr("class", "node")
          .attr("transform", function(d) { return "translate(" + d.x + "," + d.y + ")"; })
        .call(d3.behavior.drag()
          .origin(function(d) { return d; })
          .on("dragstart", function() { this.parentNode.appendChild(this); })
          .on("drag", dragmove));

      node.append("rect")
          .attr("height", function(d) { return d.dy; })
          .attr("width", sankey.nodeWidth())
          .style("fill", function(d) { return d.color = color(d.id.replace(/ .*/, "")); })
          .style("stroke", function(d) { return d3.rgb(d.color).darker(2); })
        .append("title")
          .text(function(d) { return d.id + "\nP=" + format(100.0*d.p) + "%" ;});

      node.append("text")
          .attr("x", -6)
          .attr("y", function(d) { return d.dy / 2; })
          .attr("dy", ".35em")
          .attr("text-anchor", "end")
          .attr("transform", null)
          .text(function(d) { return d.label; })
        .filter(function(d) { return d.x < _width / 2; })
          .attr("x", 6 + sankey.nodeWidth())
          .attr("text-anchor", "start");

      function dragmove(d) {
        d3.select(this).attr("transform", "translate(" + d.x + "," + (d.y = Math.max(0, Math.min(height - d.dy, d3.event.y))) + ")");
        sankey.relayout();
        link.attr("d", path);
      }
    };
    plot_sankey(graph);

}
plot_prob_graph(window.energy, '#chart');
plot_prob_graph(window.binomgraph, '#chart2');
```
